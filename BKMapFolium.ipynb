```python
import json
import folium as flm
```

```python
with open (r'C:\Users\CHerbert\Downloads\ACS2018_forwork\JSON_Practice\pluto (3).geojson') as f:
    parkslopeblock=json.load(f)

BK_coords = (40.68,-73.975) # specifying our location data
h = flm.Map(location = BK_coords, zoom_start = 18, tiles = 'cartodbpositron', zoom_control = False)

flm.GeoJson(parkslopeblock,name='brooklyn_block',overlay=True,control=True,show=True,popup=flm.features.GeoJsonPopup(fields=['address'], localize=True)).add_to(h)

h.save('BKmap6.html')

# this website has the different folium.GeoJson options but not sure if labelling by polygons by their attributes is included..
#https://python-visualization.github.io/folium/modules.html
##GeoJsontooltip function works!!
#people on github say they have more control over individual icons using 'marker' rather than GeoJsonToolTip
```
